```python
import numpy as np
from scipy.stats import chi2_contingency
import scipy.stats as stats
import pandas as pd
from scipy.stats import mannwhitneyu
```

```python
participants = pd.read_csv("participants.csv")
```

```python
grouped_diag=participants.groupby(['dataset', 'diagnosis'])
grouped_diag.describe()
```

```python
grouped_diag_sex=participants.groupby(['dataset', 'diagnosis', 'sex'])
grouped_diag_sex.describe() # 1 is female
```

```python
participants.groupby(['diagnosis']).describe()["mean_fd"]
```

```python
# Shapiro-Wilk test 
# Null hypothesis: the data is normally distributed

continuous_vars = ["age", 'mean_fd']
for var in continuous_vars:
    stat, p = stats.shapiro(participants[var])
    print(f"{var}: p={p:.4f}")
```

```python
site_contingency_tables = {}

for site in participants['dataset'].unique():
    site_data = participants[participants['dataset'] == site]
    contingency_table = pd.crosstab(site_data['diagnosis'], site_data['sex'])
    site_contingency_tables[site] = contingency_table
```

```python
for dataset in site_contingency_tables:
    chi2_stat, p_val, dof, ex = stats.chi2_contingency(site_contingency_tables[dataset])
    print(f"{dataset}: Chi2: {chi2_stat:.4f}, p={p_val:.4f}")

```

## Age comparison between groups for each site

```python
results = []
for site in participants["dataset"].unique():
    site_data = participants[participants["dataset"] == site]
    nc_age = site_data[site_data["diagnosis"] == 0]['age']
    scz_age = site_data[site_data["diagnosis"] == 1]['age']
    stat, p_value = mannwhitneyu(nc_age, scz_age, alternative='two-sided')
    results.append((site, stat, "{:.10f}".format(p_value)))

results_df = pd.DataFrame(results, columns=['Dataset', 'U', 'P-value'])
results_df
```

```python
results = []
for site in participants["dataset"].unique():
    site_data = participants[participants["dataset"] == site]
    nc_fd = site_data[site_data["diagnosis"] == 0]['mean_fd']
    scz_fd = site_data[site_data["diagnosis"] == 1]['mean_fd']
    stat, p_value = mannwhitneyu(nc_fd, scz_fd, alternative='two-sided')
    results.append((site, stat, "{:.10f}".format(p_value)))

results_df = pd.DataFrame(results, columns=['Dataset', 'U', 'P-value'])
results_df
```
